```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python

def f(x):
    return x + 2*np.sin(1.5*x)

def h(theta,X):
  val = np.matmul(X,theta)
  return val


def func(X,A):
  y_rec = []
  for i in range(0,numExp):
    noise = np.random.normal(0,2,n)
    y = f(x) + noise
    Y = y.reshape(n,1)
    theta = np.matmul(A,Y)
    ycap = h(theta,X)
    plt.plot(x,ycap)
    y_rec.append(ycap[:,0])
  y_rec = np.array(y_rec)
  plt.show()
  
  fx_fx_rec = np.square(np.tile(f(x),(numExp,1)) - y_rec)
  MSE = np.mean(fx_fx_rec,0)
  
  Ef_rec = np.mean(y_rec,0)

  bias = np.square(f(x) - Ef_rec)

  fx_rec_Ef_rec = np.square(y_rec - np.tile(Ef_rec,(numExp,1)))

  var = np.mean(fx_rec_Ef_rec,axis = 0)
  print("MSE :",np.sum(MSE))

  for i in range(0,n):
    print("bias_square + var =",bias[i] + var[i],"MSE =",MSE[i])
  print("bias, var, mean",np.sqrt(np.mean(bias)),np.mean(var),np.mean(MSE))
  plt.plot(x,f(x))
  plt.plot(x,Ef_rec)
  plt.show()

pi = np.pi
period = (2*pi)/1.5
k = 2

n = 1000
numExp = 1000

x = np.linspace(0,k*period,n)
```

```python
X1 = np.append(np.ones((n,1)),x.reshape(n,1),1)
A1 = np.matmul(np.linalg.inv(np.matmul(np.transpose(X1),X1)),np.transpose(X1))
func(X1,A1)
```
