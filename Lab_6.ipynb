```python
f = h5py.File('faceimages.mat','r')
data = f.get('data')
data = np.array(data)

X = data[0:10304,:]
# X =np.transpose(X)
plt.imshow(X[:,0].reshape(92,112).T)
```

```python
CV = np.cov(X)
[eig,U] = np.linalg.eig(CV)
```

```python
A = np.transpose(U)
Y = np.dot(A,X)
```

```python
limit = 50

Yt = Y[0:limit,:]
Ut = U[:,0:limit]
Xr = np.dot(Ut,Yt)
Xr = Xr.astype(float)
mse = np.mean((X-Xr)**2)
print("mse=",mse)
plt.imshow(Xr[:,0].reshape(92,112).T)
```
