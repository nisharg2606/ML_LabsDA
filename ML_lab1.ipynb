```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
x1 = np.linspace(-10,10,200)
var = x1*x1
plt.plot(x1,var)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x2 = np.linspace(-10,10,200)

for i in x1:
    var = i*i + x2*x2
    ax.scatter(i*np.ones(len(x2)),x2,var)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
min_i=0
min_j=0
min_err=-1
x3 = np.linspace(-25,25,100)

x4 = np.linspace(-10,10,40)
for i in x3:
    for j in x4:
        err=0
        for k in range(0,len(x)):
            err += (y[k]-(i+(j*x[k])))**2
        ax.scatter(i,j,err)
        if(min_err==-1):
            min_err=err
            min_i=i
            min_j=j
        elif err<=min_err:
            min_i=i
            min_j=j
            min_err=err
#         print(err)
plt.show()
print(min_i,min_j,min_err)
#err.subs({th0:1, th1:1})

```

```python
l=np.ones([len(x),2])
l[:,1]=x

ans=np.linalg.pinv(l).dot(y)
print(ans)
```

```python
th0= 10
th1= -5
err0=0
err=0
for k in range(0,len(x)):
    err0 += (y[k]-abs(ans[0]+(ans[1]*x[k])))**2
for k in range(0,len(x)):
    err += (y[k]-abs(th0+(th1*x[k])))**2
print(err0,err)
```
