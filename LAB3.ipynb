```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

exf = pd.read_excel(r'ex2data1-logistic.xls')
y = exf.y
x1 = (exf.x1-exf.x1.mean())/(exf.x1.std())
x2 = (exf.x2-exf.x2.mean())/(exf.x2.std())

cur1 = np.random.randint(10)+1
cur2 = np.random.randint(10)+1
cur3 = np.random.randint(10)+1
rate = 0.01
limit = 0.000001
diff1 = 1
diff2 = 1
diff3 = 1
i = 0
i_limit = 1000
m = int(np.floor(0.9*len(y)))

def df1(cur1,cur2,cur3):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it])-y[it])
    return err

def df2(cur1,cur2,cur3):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it])-y[it])*x1[it]
    return err

def df3(cur1,cur2,cur3):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it])-y[it])*x2[it]
    return err

while diff1 > limit and diff2 > limit and diff3 > limit and i < i_limit:
    prev1 = cur1
    prev2 = cur2
    prev3 = cur3
    cur1 = cur1 - rate * df1(prev1,prev2,prev3) 
    cur2 = cur2 - rate * df2(prev1,prev2,prev3)
    cur3 = cur3 - rate * df3(prev1,prev2,prev3)
    diff1 = abs(cur1 - prev1)
    diff2 = abs(cur2 - prev2)
    diff3 = abs(cur3 - prev3)
    i += 1

print(cur1,cur2,cur3)
error = 0
for it in range(m,len(y)):
    if y[it]!=np.round(sigmoid(cur1+cur2*x1[it]+cur3*x2[it])):
        error+=1
        
print(error*100/(len(y)-m),"%")
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

exf = pd.read_excel(r'ex2data2-logistic.xls')
y = exf.y
x1 = (exf.x1-exf.x1.mean())/(exf.x1.std())
x2 = (exf.x2-exf.x2.mean())/(exf.x2.std())

cur1 = np.random.randint(10)+1
cur2 = np.random.randint(10)+1
cur3 = np.random.randint(10)+1
cur4 = np.random.randint(10)+1
cur5 = np.random.randint(10)+1
rate = 0.01
limit = 0.000001
diff1 = 1
diff2 = 1
diff3 = 1
diff4 = 1
diff5 = 1
i = 0
i_limit = 1000
m = int(np.floor(0.9*len(y)))

def df1(cur1,cur2,cur3,cur4,cur5):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it]+cur4*(x1[it]*x1[it])+cur5*(x2[it]*x2[it]))-y[it])
    return err

def df2(cur1,cur2,cur3,cur4,cur5):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it]+cur4*(x1[it]*x1[it])+cur5*(x2[it]*x2[it]))-y[it])*x1[it]
    return err

def df3(cur1,cur2,cur3,cur4,cur5):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it]+cur4*(x1[it]*x1[it])+cur5*(x2[it]*x2[it]))-y[it])*x2[it]
    return err

def df4(cur1,cur2,cur3,cur4,cur5):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it]+cur4*(x1[it]*x1[it])+cur5*(x2[it]*x2[it]))-y[it])*(x1[it]*x1[it])
    return err

def df5(cur1,cur2,cur3,cur4,cur5):
    err = 0
    for it in range(0,m):
        err += (sigmoid(cur1+cur2*x1[it]+cur3*x2[it]+cur4*(x1[it]*x1[it])+cur5*(x2[it]*x2[it]))-y[it])*(x2[it]*x2[it])
    return err

while diff1 > limit and diff2 > limit and diff3 > limit and diff4 > limit and diff5 > limit and i < i_limit:
    prev1 = cur1
    prev2 = cur2
    prev3 = cur3
    prev4 = cur4
    prev5 = cur5
    cur1 = cur1 - rate * df1(prev1,prev2,prev3,prev4,prev5) 
    cur2 = cur2 - rate * df2(prev1,prev2,prev3,prev4,prev5)
    cur3 = cur3 - rate * df3(prev1,prev2,prev3,prev4,prev5)
    cur4 = cur4 - rate * df4(prev1,prev2,prev3,prev4,prev5)
    cur5 = cur5 - rate * df5(prev1,prev2,prev3,prev4,prev5)
    diff1 = abs(cur1 - prev1)
    diff2 = abs(cur2 - prev2)
    diff3 = abs(cur3 - prev3)
    diff4 = abs(cur4 - prev4)
    diff5 = abs(cur5 - prev5)
    i += 1

print(cur1,cur2,cur3,cur4,cur5)
error = 0
for it in range(m,len(y)):
    if y[it]!=np.round(sigmoid(cur1+cur2*x1[it]+cur3*x2[it]+cur4*(x1[it]*x1[it])+cur5*(x2[it]*x2[it]))):
        error+=1
        
print(error*100/abs(len(y)-m),"%")
```
